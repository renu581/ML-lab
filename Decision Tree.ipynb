## Handling Missing Values

```python
#Replace the missing values in Embarked column by 'S'
train_data['Embarked'].fillna('S').isnull().sum()
```

```python
#Impute the missing values in Fare column with median value
# Fill missing values in 'Fare' with the median value
train_data['Fare'].fillna(train_data['Fare'].median(), inplace=True)

print(train_data['Fare'].isnull().sum())  # Should print 0
```

```python
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
import numpy as np

# Calculate mean and standard deviation of non-missing Age values
age_mean = train_data['Age'].mean()
age_std = train_data['Age'].std()
num_missing = train_data['Age'].isnull().sum()

# Generate random values from a normal distribution
random_ages = np.random.normal(loc=age_mean, scale=age_std, size=num_missing)

# Fill missing Age values with these random samples
train_data.loc[train_data['Age'].isnull(), 'Age'] = random_ages

# Verify if missing values are replaced
print(train_data['Age'].isnull().sum())  # Should print 0
```

## Feature Engineering - Creating new features

```python
# Create and add a new feature FamilySize as a combination of SibSp and Parch
#FamilySize = Parch + SibSp + 1
# Create FamilySize feature
train_data['FamilySize'] = train_data['SibSp'] + train_data['Parch'] + 1

# Verify the new column
print(train_data[['SibSp', 'Parch', 'FamilySize']].head())
```

```python
# Create new feature IsAlone from FamilySize
#If FamilySize == 1 then IsAlone = True else IsAlone = False
# Create IsAlone feature
train_data['IsAlone'] = train_data['FamilySize'] == 1

# Convert Boolean to integer (0 = Not Alone, 1 = Alone)
train_data['IsAlone'] = train_data['IsAlone'].astype(int)

# Verify the new column
print(train_data[['FamilySize', 'IsAlone']].head())
```

## Create a new attribute Has_cabin using the attribue Cabin. 
#if Cabin == NaN the Has_cabin = False else Has_cabin = True


```python

#Drop the columns which may not be useful in prediction of survival
#Drop PassengerId, Name, Ticket, and Cabin
# Drop columns that are not useful for survival prediction
train_data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace=True)

# Verify the remaining columns
print(train_data.head())
```

## Model Building

```python
# Define target variable (y)
y = train_data["Survived"]  # Assuming "Survived" is the target column

# Define features (X) - Drop target and irrelevant columns
X = train_data.drop('Survived',axis=1)

# Verify shapes
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)

# Display first few rows of X and y
print(X.head())
print(y.head())
```

```python
# Label encoding of categorical features
from sklearn.preprocessing import LabelEncoder

# Identify categorical columns
categorical_columns = ['Sex', 'Embarked', 'Pclass', 'IsAlone']  # Add other categorical columns if needed

# Apply Label Encoding
label_encoder = LabelEncoder()

for col in categorical_columns:
    if col in train_data.columns:  # Ensure the column exists
        train_data[col] = label_encoder.fit_transform(train_data[col])

# Verify encoding
print("Label Encoding Successful!")
print(train_data.head())
    
```

```python
# concat df_categorical with original df

import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Find categorical columns that actually exist in train_data
categorical_columns = ['Sex', 'Embarked', 'Pclass', 'IsAlone']
existing_categorical_columns = [col for col in categorical_columns if col in train_data.columns]

# Print found categorical columns
print("Existing categorical columns:", existing_categorical_columns)

# If valid categorical columns exist
if existing_categorical_columns:
    df_categorical = train_data[existing_categorical_columns].copy()
    label_encoder = LabelEncoder()

    # Apply Label Encoding
    for col in existing_categorical_columns:
        df_categorical[col] = label_encoder.fit_transform(df_categorical[col])

    # Concatenate back to train_data
    train_data = pd.concat([train_data, df_categorical], axis=1)

    print("Encoding and concatenation successful!")
    print(train_data.head())
else:
    print("No categorical columns found in train_data. Check column names.")
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
from sklearn.model_selection import train_test_split

# Define feature matrix X and target variable y
X = train_data.drop(columns=['Survived'])  # Assuming 'Survived' is the target column
y = train_data['Survived']

# Split data into training (90%) and validation (10%) sets
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)

# Check the shape of the split datasets
print(f"X_train shape: {X_train.shape}, X_validation shape: {X_validation.shape}")
print(f"y_train shape: {y_train.shape}, y_validation shape: {y_validation.shape}")
```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier(random_state=42)
model.fit(X_train,y_train)
print(model)

from sklearrn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(random_state=42)
model
```

```python
#Print the tree

# Importing required packages for visualization
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Plot the decision tree
plt.figure(figsize=(15, 10))
plot_tree(model, feature_names=X.columns, class_names=["Not Survived", "Survived"], filled=True, rounded=True)
plt.show()
```

```python
# Check the evaluation metrics of our default model on validation set - Confusion Matrix, Accuracy, Precision, Recall, F1 score

# Importing classification report and confusion matrix from sklearn metrics

# Making predictions

# Printing classification report
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Make predictions on the validation set
y_pred = model.predict(X_validation)

# Compute confusion matrix
conf_matrix = confusion_matrix(y_validation, y_pred)

# Compute evaluation metrics
accuracy = accuracy_score(y_validation, y_pred)
precision = precision_score(y_validation, y_pred)
recall = recall_score(y_validation, y_pred)
f1 = f1_score(y_validation, y_pred)

# Print results
print("Confusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", classification_report(y_validation, y_pred))
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```
